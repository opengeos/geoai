## Import libraries

```python
import leafmap
import geoai
```

## Download sample NAIP imagery

Sample NAIP imagery: https://huggingface.co/datasets/giswqs/geospatial/tree/main/naip

```python
train_raster_url = "https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip/m_4609931_ne_14_1_20100629.tif"
test_raster_url = "https://huggingface.co/datasets/giswqs/geospatial/resolve/main/naip/m_4609932_nw_14_1_20100629.tif"
```

```python
train_raster_path = geoai.download_file(train_raster_url)
test_raster_path = geoai.download_file(test_raster_url)
```

```python
geoai.print_raster_info(test_raster_path, show_preview=False)
```

```python
geoai.view_raster(train_raster_url)
```

```python
geoai.view_raster(test_raster_url)
```

## Download sample wetland data

National Wetlands Inventory: https://www.fws.gov/program/national-wetlands-inventory/wetlands-mapper

```python
m = leafmap.Map()
m.add_raster(train_raster_path, bands=[4, 1, 2], layer_name="NAIP Imagery")
m
```

```python
bbox = m.user_roi_bounds()
if bbox is None:
    bbox = [-99.192, 46.56, -99.1206, 46.6283]
```

```python
bbox_geometry = leafmap.bbox_to_gdf(bbox)
```

```python
nwi_gdf = leafmap.get_nwi(bbox_geometry)
nwi_gdf.head()
```

```python
m.add_nwi(nwi_gdf, col_name="WETLAND_TY")
```

```python
nwi_geojson = "nwi_wetlands.geojson"
nwi_gdf.to_file(nwi_geojson)
```

## Create training data

```python
out_folder = "output"
tiles = geoai.export_geotiff_tiles(
    in_raster=train_raster_path,
    out_folder=out_folder,
    in_class_data=nwi_geojson,
    tile_size=512,
    stride=256,
    buffer_radius=0,
)
```

## Train object detection model

```python
# Train U-Net model
geoai.train_segmentation_model(
    images_dir=f"{out_folder}/images",
    labels_dir=f"{out_folder}/labels",
    output_dir=f"{out_folder}/unet_models",
    architecture="unet",
    encoder_name="resnet34",
    encoder_weights="imagenet",
    num_channels=4,
    num_classes=2,
    batch_size=8,
    num_epochs=20,
    learning_rate=0.001,
    val_split=0.2,
    verbose=True,
)
```

## Run inference

```python
masks_path = "wetland_prediction.tif"
model_path = f"{out_folder}/unet_models/best_model.pth"
```

```python
# Run semantic segmentation inference
geoai.semantic_segmentation(
    input_path=test_raster_path,
    output_path=masks_path,
    model_path=model_path,
    architecture="unet",
    encoder_name="resnet34",
    num_channels=4,
    num_classes=2,
    window_size=512,
    overlap=256,
    batch_size=8,
)
```

## Vectorize masks

```python
output_path = "wetland_prediction.geojson"
gdf = geoai.raster_to_vector(
    masks_path, output_path, min_area=1000, simplify_tolerance=0
)
```

## Visualize results

```python
geoai.view_vector_interactive(output_path, tiles=test_raster_path)
```

```python
geoai.create_split_map(
    left_layer=output_path,
    right_layer=test_raster_url,
    left_args={"style": {"color": "red", "fillOpacity": 0.4}},
    basemap=test_raster_path,
)
```
