This notebook demonstrates how to discover, filter, and load remote sensing
foundation models using the `geoai` package. The catalog is drawn from the
[Awesome Remote Sensing Foundation Models](https://github.com/Jack-bo1220/Awesome-Remote-Sensing-Foundation-Models)
curated list and integrates with [TerraTorch](https://github.com/IBM/terratorch)
for model loading and fine-tuning.

**What you will learn:**
- Browse the full catalog of remote sensing foundation models
- Filter by category, modality, task, or framework support
- Retrieve detailed metadata for any model
- Load a model backbone via TerraTorch for inference or fine-tuning

## Import Libraries

```python
import geoai
```

## List All Foundation Models

Call `geoai.list_foundation_models()` with no arguments to see the full catalog
as a pandas DataFrame. Each row is one registered model with summary metadata.

```python
df = geoai.list_foundation_models(verbose=False)
df
```

## Filter by Category

Models are organised into five categories:

| Category | Description |
|---|---|
| `vision` | Self-supervised vision-only models (MAE, contrastive, etc.) |
| `vision-language` | Models combining image and text understanding (VLM, SAM, etc.) |
| `generative` | Diffusion and generative models |
| `vision-location` | Geography- and location-aware models |
| `agents` | Agent-based earth observation systems |

```python
df_vision = geoai.list_foundation_models(category="vision", verbose=False)
df_vision
```

```python
df_vlm = geoai.list_foundation_models(category="vision-language", verbose=False)
df_vlm
```

## Filter by Modality

Filter to a specific sensor modality such as `"multispectral"`, `"sar"`,
`"hyperspectral"`, `"multimodal"`, `"timeseries"`, `"optical"`, or `"lidar"`.

```python
df_ms = geoai.list_foundation_models(modality="multispectral", verbose=False)
df_ms
```

```python
df_hyper = geoai.list_foundation_models(modality="hyperspectral", verbose=False)
df_hyper
```

## Filter by Downstream Task

Use `task=` to keep only models that list a particular downstream capability.

```python
df_seg = geoai.list_foundation_models(task="segmentation", verbose=False)
df_seg
```

```python
df_reg = geoai.list_foundation_models(task="regression", verbose=False)
df_reg
```

## TerraTorch-Supported Models

Set `terratorch_only=True` to see only models you can load directly via
`geoai.load_foundation_model()` without custom code.

```python
df_tt = geoai.list_foundation_models(terratorch_only=True, verbose=False)
df_tt
```

## Models Available on HuggingFace Hub

## Combine Filters

Filters compose вЂ” apply category, modality, task, and framework flags together.

```python
df_combined = geoai.list_foundation_models(
    category="vision",
    modality="multispectral",
    task="segmentation",
    terratorch_only=True,
    verbose=False,
)
df_combined
```

## Get Detailed Model Info

`geoai.get_foundation_model_info()` returns the full metadata dictionary
for a single model, including paper URL, code URL, license, and description.

```python
info = geoai.get_foundation_model_info("prithvi-eo-2.0-300m")
for key, value in info.items():
    print(f"{key}: {value}")
```

```python
info_clay = geoai.get_foundation_model_info("clay-v1")
for key, value in info_clay.items():
    print(f"{key}: {value}")
```

## Load a Model via TerraTorch

`geoai.load_foundation_model()` uses TerraTorch's backbone registry to
download and instantiate the model. A GPU is not required for the registry
call itself, but inference will be faster with one.

> **Note:** TerraTorch must be installed (`pip install terratorch`) and the
> model must be listed as `terratorch_supported: True` in the catalog.

## TerraTorch Backbone Keys

Each TerraTorch-supported model has an exact backbone key used internally
by `geoai.load_foundation_model()`. The table below shows the mapping.

```python
import geoai as _g

df_keys = geoai.list_foundation_models(terratorch_only=True, verbose=False)[
    ["name", "abbreviation", "modality", "publication"]
].copy()
df_keys["terratorch_key"] = [
    _g.FOUNDATION_MODELS[k]["terratorch_key"]
    for k in _g.FOUNDATION_MODELS
    if _g.FOUNDATION_MODELS[k]["terratorch_supported"]
]
df_keys
```

```python
try:
    model = geoai.load_foundation_model("prithvi-eo-2.0-300m", pretrained=True)
    n_params = sum(p.numel() for p in model.parameters())
    print(f"Model type : {type(model).__name__}")
    print(f"Parameters : {n_params:,}")
    print(f"Trainable  : {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")
    print()
    print("Model loaded successfully вЂ” ready for fine-tuning or inference.")
except ImportError as e:
    print(f"[!] TerraTorch not installed: {e}")
    print("    Fix: pip install terratorch")
except NotImplementedError as e:
    print(f"[!] Model not yet TerraTorch-supported:\n    {e}")
except RuntimeError as e:
    print(f"[!] TerraTorch registry error:\n    {e}")
```
